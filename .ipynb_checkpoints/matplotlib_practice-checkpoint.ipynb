# 5.1: Creating Visualizations Using Matplotlib

### 5.1.3: Create Line Charts

```python
import matplotlib.pyplot as plt
```

```python
# Create a Line Chart Using the MATLAB Method

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create a Line Chart Using the Object-Oriented Interface

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Use plt.show()

# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

### 5.1.4: Annotate Charts

```python
# Annotate Charts Using the MATLAB Method

# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Annotate Charts Using the Object-Oriented Method
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)

# Change the line color to green
ax.plot(x_axis,y_axis, color="green",label="Boston")
```

### 5.1.5: Create Bar Charts Using the MATLAB Approach

```python
# Create A Vertical Bar Chart

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create a Horizontal Bar Chart

# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# Create the plot
plt.barh(y_axis, x_axis)
```

```python
# Create the plot.
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Create the plot.
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
plt.title("PyBer Fare by Month")
plt.xlabel("Fare ($)")
plt.ylabel("Date")
plt.gca().invert_yaxis()
```
